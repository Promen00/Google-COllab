```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv('https://github.com/Promen00/Google-COllab/raw/main/cleaned_dataset.csv')
df
```

```python
df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
pd.to_numeric(df1.TotalCharges)
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1.dtypes
```

```python
tenure_churn_false = df1[df1.Churn=='FALSE'].tenure
tenure_churn_true = df1[df1.Churn=='TRUE'].tenure

plt.xlabel("Tenure")
plt.ylabel("Number of Customers")
plt.title("Customer Churn Prediction Visualization")
plt.hist([tenure_churn_true, tenure_churn_false], color = ['green','red'], label = ['Churn = TRUE', 'Churn = FALSE'])
plt.legend()
```

```python
tenure_churn_false = df1[df1.Churn=='FALSE'].MonthlyCharges
tenure_churn_true = df1[df1.Churn=='TRUE'].MonthlyCharges

plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")
plt.title("Customer Churn Prediction Visualization")

plt.hist([tenure_churn_true, tenure_churn_false], color = ['green','red'], label = ['Churn = TRUE', 'Churn = FALSE'])
plt.legend()
```

```python
def print_unique_col_values(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column} : {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No internet service', 'No',inplace = True)
df1.replace('FALSE', 'No',inplace = True)
df1.replace('TRUE', 'Yes',inplace = True)
df1.replace('0', 'No',inplace = True)
df1.replace('1', 'Yes',inplace = True)
df1.replace('No phone service', 'No',inplace = True)
```

```python
print_unique_col_values(df1)
```

```python
yes_no_columns = ['Partner','SeniorCitizen','Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

for col in yes_no_columns:
  df1[col].replace({'Yes': 1, 'No': 0}, inplace = True)
```

```python
for col in df1:
  print(f'{col}: {df1[col].unique()}')
```

```python
df1['gender'].replace({'Female': 1, 'Male': 0},inplace = True)
```

```python
df1['gender'].unique()
```

```python
df2 = pd.get_dummies(data = df1,columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```
