### Balancing the data

```python
df['Churn'].value_counts()
```

```python
from sklearn.utils import resample

fraud = df[df['Churn'] ==0]
not_fraud = df[df['Churn'] == 1]

balance_data = resample(fraud,
                        replace=True,
                        n_samples = len(not_fraud),
                        random_state=50)
df_balance = pd.concat([not_fraud, balance_data])
df_balance.Churn.value_counts()
```

### Shuffle the Data

```python
df_balance = df_balance.sample(frac=1)
df_balance.shape
```

### Normalization

```python
df_feature = df_balance.drop(['Churn','customerID'], axis=1)
df_label = df_balance['Churn']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df_feature)
df_feature_scaler = scaler.transform(df_feature) #return in ndarray

#convert to dataframe
df_feature_scaler = pd.DataFrame(data=df_feature_scaler,columns=df_feature.columns)
df_feature_scaler.head()
```

### Split Data

##Calculation and Evaluation Metrics For the MachineLearning Models

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

```python
# Define a function to calculate cross-validated metrics
def calculate_cv_metrics(model, X, y):
    accuracy = cross_val_score(model, X, y,scoring='accuracy')
    precision = cross_val_score(model, X, y,scoring='precision')
    recall = cross_val_score(model, X, y,scoring='recall')
    f1 = cross_val_score(model, X, y,scoring='f1')
    auc_scores = cross_val_score(model, X, y,scoring='roc_auc')
    return {'accuracy': accuracy.mean(), 'precision': precision.mean(), 'recall': recall.mean(), 'f1': f1.mean(), 'auc_scores':auc_scores.mean()}
```

```python
# Calculate cross-validated metrics for each model
rf_metrics = calculate_cv_metrics(best_rf_model, X_test, y_test)
xgb_metrics = calculate_cv_metrics(best_xgb_model, X_test, y_test)
logistic_metrics = calculate_cv_metrics(best_logistic_model, X_test, y_test)
```

#DeepLearning Models

```python
import tensorflow as tf
hidden_layer_size = 150
output_size = 2 #0 and 1

#3 hidden layer, depth 100
model = tf.keras.Sequential([
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(output_size, activation='softmax')
])

#optimizer
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)
history = model.fit(train_inputs,
          train_targets,
          batch_size = 200,
          epochs = 15,
          callbacks = [early_stopping],
          validation_data = (val_inputs, val_targets),
          verbose = 2)
```

```python
def plot_learningCurve(history):
    # Extracting training history
    epochs = len(history.history['accuracy'])
    epoch_range = range(1, epochs + 1)
    # Plotting training and validation accuracy
    plt.plot(epoch_range, history.history['accuracy'], label='Training Accuracy')
    plt.plot(epoch_range, history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
plot_learningCurve(history)
```

```python
test_loss, test_accuracy = model.evaluate(test_inputs, test_targets)
print("Test loss: {0:.2f} and Test accuracy: {1:.2f}% ".format(test_loss,test_accuracy*100))
```

## Deep Learning Classification Using LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Reshape data for LSTM input
train_inputs_lstm = train_inputs.reshape(train_inputs.shape[0], 1, train_inputs.shape[1])
val_inputs_lstm = val_inputs.reshape(val_inputs.shape[0], 1, val_inputs.shape[1])
test_inputs_lstm = test_inputs.reshape(test_inputs.shape[0], 1, test_inputs.shape[1])
```

```python
# Define LSTM model
lstm_model = Sequential([
    LSTM(hidden_layer_size, input_shape=(train_inputs_lstm.shape[1], train_inputs_lstm.shape[2])),
    Dense(output_size, activation='softmax')
])
```

```python
# Compile LSTM model
lstm_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train LSTM model
early_stopping_lstm = tf.keras.callbacks.EarlyStopping(patience=2)
history_lstm = lstm_model.fit(train_inputs_lstm,
                              train_targets,
                              batch_size=200,
                              epochs=15,
                              callbacks=[early_stopping_lstm],
                              validation_data=(val_inputs_lstm, val_targets),
                              verbose=2)
```

```python
# Plot learning curve for LSTM model
plot_learningCurve(history_lstm)
```

```python
# Evaluate LSTM model
test_loss_lstm, test_accuracy_lstm = lstm_model.evaluate(test_inputs_lstm, test_targets)
print("LSTM Test loss: {0:.2f} and Test accuracy: {1:.2f}% ".format(test_loss_lstm, test_accuracy_lstm*100))
```

## PARALLEL LSTM-CNN ARCHITECTURE

```python
from tensorflow.keras.layers import Input, LSTM, Dense, Conv1D, GlobalMaxPooling1D, Concatenate, ZeroPadding1D
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dropout
from tensorflow.keras.regularizers import l1_l2

# Input layer
train_inputs_lstm_shape = train_inputs_lstm.shape
input_shape = (train_inputs_lstm_shape[1], train_inputs_lstm_shape[2])
input_layer = Input(shape=input_shape, name='input')
```

```python
# LSTM branch
lstm_branch = LSTM(hidden_layer_size, dropout=0.2, recurrent_dropout=0.2)(input_layer)
```

```python
# Add ZeroPadding1D layer to increase input size for CNN branch
cnn_input_padded = ZeroPadding1D(padding=1)(input_layer)

# CNN branch with dropout
cnn_branch = Conv1D(64, 3, activation='relu')(cnn_input_padded)
cnn_branch = GlobalMaxPooling1D()(cnn_branch)  # Global max pooling
cnn_branch = Dense(128, activation='relu')(cnn_branch)
cnn_branch = Dropout(0.2)(cnn_branch)
```

```python
# Concatenate LSTM and CNN outputs
merged_output = Concatenate()([lstm_branch, cnn_branch])
```

```python
# Joint output layer
output = Dense(output_size, activation='softmax', kernel_regularizer=l1_l2(l1=0.001, l2=0.001))(merged_output)
```

```python
# Create parallel LSTM-CNN model
parallel_model = Model(inputs=input_layer, outputs=output)
```

```python
# Compile the model
parallel_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
early_stopping_parallel = tf.keras.callbacks.EarlyStopping(patience=2)
history_parallel = parallel_model.fit(train_inputs_lstm,
                                      train_targets,
                                      batch_size=200,
                                      epochs=15,
                                      callbacks=[early_stopping_parallel],
                                      validation_data=(val_inputs_lstm, val_targets),
                                      verbose=2)
```

```python
# Plot learning curve for the parallel model
plot_learningCurve(history_parallel)
```

```python
# Evaluate the parallel LSTM-CNN model
test_loss_parallel, test_accuracy_parallel = parallel_model.evaluate(test_inputs, test_targets)
print("Parallel LSTM-CNN Model Test loss: {0:.2f} and Test accuracy: {1:.2f}% ".format(test_loss_parallel, test_accuracy_parallel*100))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Calculate predictions for the parallel LSTM-CNN model
predictions_parallel = parallel_model.predict(test_inputs_lstm)
predicted_labels_parallel = predictions_parallel.argmax(axis=1)

# Calculate evaluation metrics for the parallel LSTM-CNN model
accuracy_parallel = accuracy_score(test_targets, predicted_labels_parallel)
precision_parallel = precision_score(test_targets, predicted_labels_parallel)
recall_parallel = recall_score(test_targets, predicted_labels_parallel)
f1_parallel = f1_score(test_targets, predicted_labels_parallel)

# Calculate confusion matrix for the parallel LSTM-CNN model
conf_matrix_parallel = confusion_matrix(test_targets, predicted_labels_parallel)

# Create a DataFrame for the parallel LSTM-CNN metrics
parallel_metrics_data = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1 Score'],
    'Value': [accuracy_parallel, precision_parallel, recall_parallel, f1_parallel]
}
parallel_metrics_df = pd.DataFrame(parallel_metrics_data)

# Create a DataFrame for the confusion matrix of the parallel LSTM-CNN model
conf_matrix_parallel_df = pd.DataFrame(conf_matrix_parallel, columns=['Predicted No Churn', 'Predicted Churn'],
                                       index=['Actual No Churn', 'Actual Churn'])

# Plotting
plt.figure(figsize=(18, 8))

# Plot metrics for the parallel LSTM-CNN model
plt.subplot(1, 3, 1)
sns.barplot(x='Value', y='Metric', data=parallel_metrics_df, palette='viridis')
plt.title('Parallel LSTM-CNN Metrics')

# Plot confusion matrix for the parallel LSTM-CNN model
plt.subplot(1, 3, 2)
sns.heatmap(conf_matrix_parallel_df, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={'size': 14})
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Parallel LSTM-CNN Confusion Matrix')

# Plot metrics for the parallel LSTM-CNN model
plt.subplot(1, 3, 1)
ax = sns.barplot(x='Value', y='Metric', data=parallel_metrics_df, palette='viridis')

# Add annotations with precision values
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='center', va='center', fontsize=12, color='black', xytext=(5, 0),
                textcoords='offset points')

plt.title('Parallel LSTM-CNN Metrics')
plt.xlabel('Value')
plt.ylabel('Metric')
plt.tight_layout()
plt.show()


# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
